```python
# convert North Wales FOI to csv

import pdfplumber
import re 
# import regex here as the table on the PDF does not follow a traditional structure (each cell with borders)
import pandas as pd

pdf_path = "/Users/c23093869/Desktop/Data Sheet DPO000341.pdf"

all_data = []

with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):  # loop through all pages
        text = page.extract_text()
        lines = text.split("\n")
        
        # Skip header only on the first page
        start_line = 1 if i == 0 else 0
        
        for line in lines[start_line:]:
            # Regex to match the structure of your data
            match = re.match(
                r'\d+\s+([A-Za-z ]+)\s+\d{2}/\d{2}/\d{4}\s+\d{2}:\d{2}\s+(\d{4})\s+([A-Za-z]+)\s+(.+)\s+\d{2}:\d{2}:\d{2}',
                line
            )
            if match:
                authority = match.group(1).strip()
                year = match.group(2).strip()
                month = match.group(3).strip()
                fire_cause = match.group(4).strip()
                all_data.append([authority, year, month, fire_cause])

# Create a pandas DataFrame
df = pd.DataFrame(all_data, columns=['Unitary Authority', 'Year', 'Month', 'Fire Cause'])

# Display the first few rows
print(df.head())

# Save to CSV
df.to_csv("/Users/c23093869/Desktop/fire_data.csv", index=False)
```

```python
import pandas as pd

# Month name to number mapping
month_to_num = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}

# Add a numeric month column
df['Month_Num'] = df['Month'].map(month_to_num)

# Filter rows from Sept 2024 to June 2025
df_filtered = df[
    ((df['Year'].astype(int) == 2024) & (df['Month_Num'] >= 9)) |
    ((df['Year'].astype(int) == 2025) & (df['Month_Num'] <= 6))
]

# Count deliberate vs accidental
deliberate_count = df_filtered['Fire Cause'].str.contains('Deliberate', case=False).sum()
accidental_count = df_filtered['Fire Cause'].str.contains('Accidental', case=False).sum()

print(f"From Sept 2024 to June 2025:")
print(f"Deliberate fires: {deliberate_count}")
print(f"Accidental fires: {accidental_count}")
```
